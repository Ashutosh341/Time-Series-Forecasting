```python

import pandas as pd
from datetime import datetime
import yfinance as yf
```

```python
df_tesla = yf.download('TSLA')
df_tesla
```

```python
## Simple Moving Average
```

```python
df_tesla['Open:10 Days Rolling'] = df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10 Days Rolling']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
df_tesla['Open:30 Days Rolling'] = df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla['Open:50 Days Rolling'] = df_tesla['Open'].rolling(window=50,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10 Days Rolling','Open:30 Days Rolling','Open:50 Days Rolling']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
## Cumulative Moving Average
```

```python
df_tesla['Open'].expanding().mean().plot()
```

```python
# Exponential  Moving Average
```

```python
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1,adjust = False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.3,adjust = False).mean()
df_tesla[['Open','EMA_0.1','EMA_0.3']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```

```python
# Exponential Weighted Moving Average
```

```python
df_tesla['EWMA_5 Days'] = df_tesla['Open'].ewm(span=5).mean()
df_tesla[['Open','EWMA_5 Days']].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,5))
```
