```python
## Time Resampling
```

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
type(df_tesla.resample(rule='A').max())
```

```python
# Year end frequency
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
# Quartely end frequency
df_tesla.resample(rule='QS').max()
```

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
plt.tight_layout()
df_tesla.iloc[:1209,:1].resample(rule='M').mean().plot(kind='bar',figsize=(15,6))
```
